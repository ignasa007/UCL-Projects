```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
# TODO: make xticks and yticks the same in each plot, make colorbar ticks the same

def plot_contour(low, high, grid_size, index, levels=20):
    f = lambda x, y: np.square(y+np.log(x)) + np.square(y-x)
    x, y = np.meshgrid(np.linspace(max(1e-1, low), high, 1000), np.linspace(low, high, 1000))
    plt.subplot(*grid_size, index)
    z = f(x, y)
    cs = plt.contourf(x, y, z, np.logspace(np.log10(z.min()),np.log10(z.max()), levels), cmap='RdYlBu', locator=ticker.LogLocator())
    cbar = plt.colorbar(cs)
    cbar.locator = ticker.LogLocator(10)
    # cbar.set_ticks(cbar.locator.tick_values(z.min(), z.max()))
    cbar.minorticks_off()
    plt.xticks(rotation=45)

plt.figure(figsize=(11, 5))
grid_size = (1, 2)
plot_contour(0.1, 10, grid_size, 1)
plot_contour(0.1, 1., grid_size, 2)
# plot_contour(0.3, 0.8, grid_size, 3)
# plot_contour(0.55, 0.58, grid_size, 4)

plt.tight_layout()
plt.show()
```
